Changes to make to the dataset:

1. Change the 'Sex' column to the following: 'M':0, 'F': 1
2. Find the outliers in `Fare` and set the high-end outliers to be equal to the 75th percentile of the `Fare` portion of the dataset.

```python
# Implementing step 1
sex_dict = {"male": 0, "female": 1}
data['Sex'] = data['Sex'].replace(sex_dict)

# Implementing step 2
percentile = 31.0
data.loc[data['Fare'] >= percentile, 'Fare'] = percentile
```
