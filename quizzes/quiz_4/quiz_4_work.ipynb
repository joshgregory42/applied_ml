```python
import numpy as np
import itertools

# Define the possible values for w_1, w_2, and b
values = [-2, -1, 0, 1, 2]

# Generate all combinations of w_1, w_2, and b
combinations = itertools.product(values, repeat=3)

# Print out each combination in a row
for w_1, w_2, b in combinations:
    row_1 = 1 if (0*w_1 + 0*w_2 + b) > 0 else 0
    row_2 = 1 if (0*w_1 + 1*w_2 + b) > 0 else 0
    row_3 = 1 if (1*w_1 + 0*w_2 + b) > 0 else 0
    row_4 = 1 if (1*w_1 + 1*w_2 + b) > 0 else 0

    print(f'Row 1: {row_1}')
    print(f'Row 2: {row_2}')
    print(f'Row 3: {row_3}')
    print(f'Row 4: {row_4}')
    print(f'w_1:{w_1}, w_2:{w_2}, b:{b}')
    print()
```

In the output above, we can see that there are corresponding weights and biases for every gate except for the XOR gate.
